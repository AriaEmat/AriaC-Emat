## English Stats

```python
df.loc[
    (df.index >= 1) & (df.index <= 575) &
    (df['Primary Spoken Language'] == 'English'),
    ['Year', 'Primary Spoken Language', ' Number of Applicants ']
]
```

```python
df[" Number of Applicants "] = (
    df[" Number of Applicants "]
    .astype(str)
    .str.strip()
    .str.replace(",", "", regex=False)
)

df[" Number of Applicants "] = pd.to_numeric(df[" Number of Applicants "], errors="coerce")
```

```python
filtered = df[df["Primary Spoken Language"] == "English"]

weekly_avg = filtered.groupby("Year")[" Number of Applicants "].mean()

print(weekly_avg)
```

```python
weekly_avg.plot(kind="bar", figsize=(8,4), color="coral")
plt.title("Average Applicants per year (English)")
plt.xlabel("Year")
plt.ylabel("# of Applicants")
plt.show()
```
